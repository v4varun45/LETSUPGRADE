```python
import numpy as np
import pandas as pd
import re
import nltk
import seaborn as sns
```

```python
dataset=pd.read_csv('Restaurant_Reviews.tsv',delimiter='\t',quoting=3)
```

```python
dataset['Review'][0]
```

```python
from nltk.stem.porter import PorterStemmer
corpus=[]
for i in range(0,1000):
    review=re.sub('[^a-zA-Z]',' ',dataset['Review'][i])
    review.lower()
    x=review.split()
    cut=PorterStemmer()
    a=[ cut.stem(word) for word in x if not word in set(stopwords.words('english'))]
    a=' '.join(a)
    print(a)
    corpus.append(a)
```

```python
corpus
```

```python
corpus_dataset=pd.DataFrame(corpus)
```

```python
corpus_dataset
```

```python
corpus_dataset['Review']=corpus_dataset
```

```python
corpus_dataset
```

```python
corpus_dataset.drop([0],axis=1)
```

```python
corpus_dataset.to_csv('corpus_dataset.csv')
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=1500)
```

```python
X=cv.fit_transform(corpus).toarray()
X
```

```python
y=dataset.iloc[:,1].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
clf=GaussianNB()
clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
input=['This place is Awesome']
op=cv.transform(input).toarray()
op=clf.predict(op)
op[0]
if op[0]==1:
    print('Review is Positive')
else:
    print('Review is Negative')
```
