# Logistic regression on BANK LOAN Dataset

```python
import pandas as pd
```

```python
pwd
```

```python
data=pd.read_excel('Bank_Personal_Loan_Modelling.xlsx',sheet_name=1)
```

```python
data.head()
```

```python
data.drop(['ID','ZIP Code'],axis=1,inplace=True)
```

```python
data.head()
```

```python
X=data[['Age','Experience','Income','Family','CCAvg','Education','Mortgage','Securities Account','CD Account','Online','CreditCard']]
```

```python
X
```

```python
Y=data['Personal Loan']
```

```python
import statsmodels.api as sm
```

```python
X1=sm.add_constant(X)
```

```python
Logistic=sm.Logit(Y,X1)
```

```python
result=Logistic.fit()
```

```python
result.summary()
```

# In the above result except Age,Experience,Mortgage all other variables are significantly important.
# To get "Personal_lone"   Income,Family,Education,CD Account,Online,CreditCard are most significantly important.

# Logistic regression on ATTRITION Dataset

```python
pwd
```

```python
import statsmodels.api as sm
```

```python
data1=pd.read_csv('general_data.csv')
```

```python
data1.head()
```

```python
from sklearn import preprocessing
```

```python
label=preprocessing.LabelEncoder()
```

```python
import numpy as np
```

```python
data1['JobRole']=label.fit_transform(data1['JobRole'])
data1['Attrition']=label.fit_transform(data1['Attrition'])
data1['BusinessTravel']=label.fit_transform(data1['BusinessTravel'])
data1['Department']=label.fit_transform(data1['Department'])
data1['EducationField']=label.fit_transform(data1['EducationField'])
data1['Gender']=label.fit_transform(data1['Gender'])
data1['MaritalStatus']=label.fit_transform(data1['MaritalStatus'])

```

```python
data=data1.dropna()
```

```python
data2=data.drop_duplicates()
```

```python
Y=data2['Attrition']
```

```python
X=data2[['Age','BusinessTravel','Department','DistanceFromHome','Education','EducationField','Gender','JobLevel','JobRole','MaritalStatus','MonthlyIncome','NumCompaniesWorked','PercentSalaryHike','StockOptionLevel','TotalWorkingYears','TrainingTimesLastYear','YearsAtCompany','YearsSinceLastPromotion','YearsWithCurrManager']]
```

```python
X1=sm.add_constant(X)
```

```python
Logistic=sm.Logit(Y,X1)
```

```python
result=Logistic.fit()
```

```python
result.summary()
```

# Linear regression with one dependent variable and one independent variable

```python
import matplotlib as plt
import statsmodels.formula.api as sm
```

```python
data2=pd.read_excel('Linear Regression.xlsx',sheet_name=0)
```

```python
data2.head()
```

```python
data2.plot(kind='scatter',x='sqft_living',y='price')
```

```python
lm=sm.ols(formula='price~sqft_living',data=data2).fit()
```

```python
lm.summary()
```

# Linear regression with one dependent variable and more than one independent variable

```python
lm=sm.ols(formula='price~sqft_living+bedrooms+bathrooms+floors',data=data2).fit()
```

```python
lm.summary()
```
